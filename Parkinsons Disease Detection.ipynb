## 3.Model Building and Evaluation

### 3.1Split the data into training and testing sets

```python
X=df.drop(labels=['status'],axis=1)#group of attributes to determine disease
display (X.head())
```

```python
Y=df['status'] #target attribute that says disease is there are not.
display (Y.head())
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=40)
print (X.shape,Y.shape)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

### 3.2Appropriate Machine Learning Algorithms  and Evaluation

## Logistic Regression

```python
# Train the Logistic Regression model and obtain predictions on the train set
log_reg = LogisticRegression().fit(X_train, Y_train)
train_preds = log_reg.predict(X_train)

# Calculate and print the accuracy on the train set
print("Model accuracy on train is: ", accuracy_score(Y_train, train_preds))

# Obtain predictions on the test set
test_preds = log_reg.predict(X_test)

# Calculate and print the accuracy on the test set
print("Model accuracy on test is: ", accuracy_score(Y_test, test_preds))
print('-' * 50)

# Compute and print the confusion matrix for the train set
print("Confusion matrix for train is:\n", confusion_matrix(Y_train, train_preds))

# Compute and print the confusion matrix for the test set
print("Confusion matrix for test is:\n", confusion_matrix(Y_test, test_preds))

# Print the classification report for the train set
print('\nClassification Report for Train:')
print(classification_report(Y_train, train_preds))

# Print the classification report for the test set
print('\nClassification Report for Test:')
print(classification_report(Y_test, test_preds))
```

## Random Forest Algorithm

```python
# Train the Random Forest classifier and obtain predictions on the train set
RF = RandomForestClassifier().fit(X_train, Y_train)
train_preds2 = RF.predict(X_train)

# Calculate and print the accuracy on the train set
print("Model accuracy on train is: ", accuracy_score(Y_train, train_preds2))

# Obtain predictions on the test set
test_preds2 = RF.predict(X_test)

# Calculate and print the accuracy on the test set
print("Model accuracy on test is: ", accuracy_score(Y_test, test_preds2))

# Compute and print the confusion matrix for the train set
print("Confusion matrix for train is:\n", confusion_matrix(Y_train, train_preds2))

# Compute and print the confusion matrix for the test set
print("Confusion matrix for test is:\n", confusion_matrix(Y_test, test_preds2))

print('Wrong predictions out of total')
print('-'*50)
print((Y_test !=test_preds2).sum(),'/',((Y_test == test_preds2).sum()+(Y_test != test_preds2).sum()))
print('-'*50)

# Print the classification report for the train set
print('\nClassification Report for Train:')
print(classification_report(Y_train, train_preds2))

# Print the classification report for the test set
print('\nClassification Report for Test:')
print(classification_report(Y_test, test_preds2))
```

#### Kappa Score

The Kappa score takes into account the agreement between the observed and expected agreements, taking into consideration the possibility of the agreement occurring by chance. It is a normalized metric that ranges from -1 to 1, where:

A Kappa score of 1 indicates perfect agreement between the raters or model predictions.
A Kappa score of 0 indicates agreement that is no better than chance.
A Kappa score less than 0 indicates agreement that is worse than chance.

```python
# Calculate the Kappa score between the true values (Y_test) and predicted values (test_preds3)
kappa_score = metrics.cohen_kappa_score(Y_test, test_preds3)

# Print the Kappa score
print('Kappa Score is:', kappa_score)
```
